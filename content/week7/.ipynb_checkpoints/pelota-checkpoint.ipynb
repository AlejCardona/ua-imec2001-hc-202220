```julia
using DifferentialEquations
using Plots
```

```julia
#=
Paso 1. Definir el problema
---
ddu = v'
du = v
u = x(t)
=#
function pelota!(du,u,p,t)
    if u > 0
        ddu = (-bₐ*du - m*g) / m # Otra forma -> ddu = (-p[1]*du - p[2]*p[3]) / p[2]
    else # x(t) <= 0
        ddu = (-k*u -b*du - m*g) / m # Otra forma -> ddu = (-p[4]*u -p[5]*du - p[2]*p[3]) / p[2]
    end
    
    if isapprox(u, 0)
        du0 = du
        u0 = u
    end
    return ddu
end

# Constantes
bₐ = 0.1
m = 1.0
b = 30.0
g = 9.8
k = 201512.0 / 2.0

# Parámetros
du0 = 0.0 # v(0) = 0
u0 = 10.0 # u(0) = 10
tspan = (0.0, 10.0)
p = [bₐ, m, g, k] # Si se utiliza 'otra forma'

# Definición
prob = SecondOrderODEProblem(pelota!, du0, u0, tspan, p)
```

```julia
# Paso 2. Resolver el problema
sol = solve(prob)
sol
```

```julia
# Paso 3. Analizar la solución
plot(sol,
     linewidth=2,
     xlabel="Tiempo (s)",
     ylabel=["v (m/s)" "x (m)"],
     color=["blue" "orange"],
     legend=false,
     layout=(2,1))
```

```julia
du_sol = [u[1] for u in sol.u]
u_sol = [u[2] for u in sol.u]

# Chequear x(t₁) = 0
count = 1
for i in 1:length(u_sol)
    if isapprox(u_sol[i], 0; atol=0.01) == true
        println("X=0 se da en t($count)=$(round(sol.t[i]; digits=2)) seg. -> i = $(i)")
        count += 1
    end
end
```
